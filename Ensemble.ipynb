```
#import and show image
image = cv2.imread('/content/sample_data/para1.jpeg')
print('Original Image')
cv2_imshow(image)
```

```
#grayscale
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
#binary
ret,thresh = cv2.threshold(gray,127,255,cv2.THRESH_BINARY_INV)
#dilation
kernel = np.ones((5,100), np.uint8)
img_dilation = cv2.dilate(thresh, kernel, iterations=1)
#find contours
im2,ctrs,hie = cv2.findContours(img_dilation.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
#sort contours
sorted_ctrs = sorted(ctrs, key=lambda ctr: cv2.boundingRect(ctr)[0])
p=len(sorted_ctrs)
```

```
arr=[]
for i, ctr in enumerate(sorted_ctrs):
    
    # Get bounding box
    x, y, w, h = cv2.boundingRect(ctr)

    # Getting ROI
    roi = image[y:y+h, x:x+w]
    arr.append(roi)


    # show ROI
    #cv2_imshow(roi)
    #cv2.rectangle(image,(x,y),( x + w, y + h ),(90,0,255),2)
    #cv2.imwrite('/content/drive/My Drive/to predict lines/%d.png'%p, roi)
    
    #cv2_imshow(roi)
    
```
